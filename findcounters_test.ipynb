## 测试一下用findContours方法检测出来的边缘有什么不同
2018年11月9日

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
src = cv2.imread('../../data/book08.jpg')
imgray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
print(type(ret),ret)
print(type(thresh),len(thresh),thresh.shape)
cv2.imshow('imgray',imgray)
#cv2.imshow('thresh',thresh)
#plt.imshow(src)
plt.imshow(thresh,cmap='gray')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
plt.imshow(image,cmap='gray')
plt.show()
print(type(image))
print(type(contours),len(contours))
print(type(hierarchy))
```

```python
for i in range(0,len(contours)): 
  x, y, w, h = cv2.boundingRect(contours[i])  
  cv2.rectangle(src, (x,y), (x+w,y+h), (153,153,0), 5)
plt.imshow(src)
plt.show()
```

### 小结：
findcontours似乎对分散度较大的图像的轮廓识别率较高，但是对于图书这样紧密排列的情况效果不好
